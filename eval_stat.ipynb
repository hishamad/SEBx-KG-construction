```python
import pandas as pd
from IPython.display import display
import numpy as np
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
eval = pd.read_csv('evaluations/evaluation_consensus.csv')
display(eval)
```

```python
n_triplets = len(eval)
_,_, correct_triplet, correct_entities , correct_relation  = np.sum(eval)
print('Number  Triplets:       ' + str(n_triplets))
print('Correct Triplets:       ' + str(correct_triplet)   + '    ' + str(correct_triplet/n_triplets * 100) + ' %')
print('Correct Entities:       ' + str(correct_entities)  + '    ' + str(correct_entities/(n_triplets*2) * 100) + ' %')
print('Correct Relation:       ' + str(correct_relation)  + '    ' + str(correct_relation/n_triplets * 100) + ' %')
```
